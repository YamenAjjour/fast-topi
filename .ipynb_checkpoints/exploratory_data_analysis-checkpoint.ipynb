```python
import pandas as pd
import re
import yaml
from nltk import ngrams
from nltk.corpus import stopwords
def load_config():
    with open('config.yaml') as file:
        config= yaml.safe_load(file)
        return config
    
def get_descriptive_category(category):
    if category == "b":
        return "business"
    elif category =="m":
        return "health"
    elif category =="t":
        return "tech"
    elif category =="e":
        return "entertainment"
    else:
        raise ValueError("Unrecognized label")
        

def load_dataset():
    config =load_config()
    path_dataset =  config['dataset']['path_source']
    df= pd.read_csv(path_dataset,sep="\t",encoding="utf-8",names=["id","title","url","publisher","category","story","hostname","timestamp"])
    df['category']=df['category'].apply(get_descriptive_category,)
    df['title']=df['title'].apply(lambda title: title.lower())
    
    return df
def tokenize_alpha(title):
    return re.findall("[a-zA-Z]+(?:'[a-zA-Z])?",title)

def tokenize_alphanummeric(title):
    return re.findall("\w+(?:'\w)?",title)

def drop_stop_words(title):
    stop_words=stopwords.words('english')
    return [word for word in title if word not in stop_words]


def get_tokens(tokenizer,no_stop_words):

    dataset_df= load_dataset()
    for _,record in dataset_df.iterrows():
        tokens= tokenizer(record['title'])
        if no_stop_words:
            yield drop_stop_words(tokens),record['category']
        else:
            yield tokens,record['category']
```

## Ngrams

```python
def create_two_grams(title):
    return [gram[0]+"_"+gram[1] for gram in ngrams(title,2)]
    
def create_three_grams(title):
    return [gram[0]+"_"+gram[1]+"_"+gram[2] for gram in ngrams(title,3)]


def get_grams(n,is_alphanummeric=True,drop_stop_words=True):
    all_grams=[]
    all_categories=[]
    if is_alphanummeric:
        tokenizer= tokenize_alpha
    else:
        tokenizer = tokenize_alphanummeric
        
    for title_tokens,category in get_tokens(tokenizer,drop_stop_words):
        if n==1:
            grams = title_tokens
        elif n==2:
            grams = create_two_grams(title_tokens)
        else:
            grams = create_three_grams (title_tokens)          
        all_grams.extend(grams)
        all_categories.extend([category for _ in grams])
    return pd.DataFrame({f"{n}-gram":all_grams,"category":all_categories})
```
